```python
# ----------------------------------------- data partition  -----------------------------------------
from sklearn.model_selection import train_test_split
print(x.shape)
print(classid.shape)
train_x,test_x,train_y,test_y=train_test_split(x,classid,shuffle=True,test_size=0.25,random_state=42)
print(train_x.shape)
print(test_x.shape)

print(train_y.shape)
print(test_y.shape)

print(test_y)
```

```python
#%%timeit 
# ----------------------------------------- use knn for classification -----------------------------------------

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)

knn.fit(train_x,train_y)

accuracy=knn.score(test_x,test_y)
print('accuracy in knn(n=3) is : '+str(accuracy))
#q=knn.predict([[0,1,0,1,0,202500,1193580,35028,855000,2,0,-3,0.025164,-17482,-1262,-1182,-1029,1,1,1,0,1]])
#print(q)
```
